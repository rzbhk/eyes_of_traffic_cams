```python
from ast import literal_eval
```

```python
res = literal_eval(a[1])
```

```python
for len_, mse_per_len in res.items():
    t_ = []
    for mses_per_trial in mse_per_len:
        t_.append(np.mean(mses_per_trial))
    print("{}, {}, {}".format(len_,np.mean(t_),np.min(t_))) #average of mean of trials and min of mean trials (best trial)
```
